# Oriented FAST and Rotated BRIEF (ORB) Matching

In this project, I took two pictures called sim0 and sim1  (using my mobile phone) with a slight change in orientation/displacement and to, I took another two pictures called lim0 and lim1 with moderate change in orientation/displacement. Then, I applied  ORB matching on the pair of sim0 and sim0, the pair sim0 and sim1, and the pair lim0 and lim1.


#Import Libraries

```python
from PIL import Image
import numpy as np
from skimage.util import img_as_float
from skimage import io
import matplotlib.pyplot as plt
import argparse
from skimage.feature import ORB, match_descriptors
```

#Load Images

```python
# load sim0 and sim1 and convert them to one dimensional array
# sim0 and sim1 are a pair with slight change in orientation/displacement
sim0 = np.array(Image.open('/content/drive/MyDrive/Colab Notebooks/sim0.jpg').convert('L'))
sim1 = np.array(Image.open('/content/drive/MyDrive/Colab Notebooks/sim1.jpg').convert('L'))

# load lim0 and lim1 and convert them to one dimensional array
# lim0 and lim1 are a pair with moderate change in orientation/displacement
lim0 = np.array(Image.open('/content/drive/MyDrive/Colab Notebooks/lim0.jpg').convert('L'))
lim1 = np.array(Image.open('/content/drive/MyDrive/Colab Notebooks/lim1.jpg').convert('L'))
```

# Apply ORB Matching on the pair of the same image sim0 and sim0



```python
# create detectors. Key points for matching between two image was selected as 100
detector_extractor0 = ORB(n_keypoints=100)
detector_extractor1 = ORB(n_keypoints=100)

# connect detectors to images
detector_extractor0.detect_and_extract(sim0)
detector_extractor1.detect_and_extract(sim0)

# find matching pairs of keypoint descriptor
matches = match_descriptors(
    detector_extractor0.descriptors,
    detector_extractor1.descriptors
)
```

```python
# define functions
def plot_matches(im0, im1, pts0, pts1, matches):
    # side by side
    image = np.concatenate((im0, im1), axis=1)

    # a copy below without matches
    image = np.vstack((image, image))
    plt.imshow(image)

    # offset
    cols0 = im0.shape[1]
    for i,m in matches: # i is for list0, m is for list1
        # draw line joining the matched pair of points

        plt.plot([pts0[i][1],pts1[m][1]+cols0], [pts0[i][0],pts1[m][0]], 'c') # same row
    plt.axis('off')
```

```python
# do plots
plt.figure()
plt.gray()

# call function
plot_matches(sim0, sim0, detector_extractor0.keypoints, detector_extractor1.keypoints, matches)
```
